```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
from math import pi
#fungsi sinus
x = np.arange(0, 5*pi, 0.1)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
#fungsi cosinus
x = np.arange(0, 5*pi, 0.1)
y = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
#membuat lingkaran
jari2 = 5

sudut = np.linspace(0, 2*pi, 100)
x = jari2*np.cos(sudut)
y = jari2*np.sin(sudut)

plt.plot(x,y)
plt.show()
```
